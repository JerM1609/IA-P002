```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import utils
import DBScan as db
import multiprocessing as mp
```

```python
IMG_PATH = 'data/'
```

```python
data = utils.get_data(IMG_PATH)
```

```python
epsilons = [0.1, 0.5, 1, 2, 5]
thresholds = [5, 10, 25, 75, 100]
```

```python
fig, ax= plt.subplots(len(epsilons), len(thresholds), figsize=(4*len(epsilons),10))


for ep in range(len(epsilons)):
    for thr in range(len(thresholds)):
        DBScan = db.DBScan(eps = epsilons[ep], threshold = thresholds[thr])
        clusters = DBScan.execute(data[0][1][['R','G','B']])
        df = data[0][1].copy()
        df['cluster'] = clusters
        ax[ep,thr].scatter(df['i'],df['j'],c=df['cluster'])
        ax[ep,thr].axis('off')
        ax[ep,thr].set_title(f"EP: {epsilons[ep]}, Threshold: {thresholds[thr]}")
        
plt.savefig('Resultados/DBScan/Cross_Validator_Params.png', bbox_inches='tight')
```
