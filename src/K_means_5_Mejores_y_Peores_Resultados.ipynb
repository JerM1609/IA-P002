```python
import pandas as pd
import numpy as np
import utils
import matplotlib.pyplot as plt
from natsort import natsorted
import os
```

## Mejores Segmentaciones tumorales

```python
PATH = 'Resultados/K_Means/Objetos_Clusterizados/'
lista_archivos = natsorted(os.listdir(PATH))

df_0 = pd.read_csv(PATH + lista_archivos[5], index_col= 0)
df_1 = pd.read_csv(PATH + lista_archivos[9], index_col= 0)
df_2 = pd.read_csv(PATH + lista_archivos[10], index_col= 0)
df_3 = pd.read_csv(PATH + lista_archivos[17], index_col= 0)


fig, ax= plt.subplots(1, 4, figsize=(20,5))

for i in range(4):
    ax[i].axis('off')
    
ax[0].scatter(df_0['i'],df_0['j'],c=df_0['cluster'])
ax[1].scatter(df_1['i'],df_1['j'],c=df_1['cluster'])
ax[2].scatter(df_2['i'],df_2['j'],c=df_2['cluster'])
ax[3].scatter(df_3['i'],df_3['j'],c=df_3['cluster'])

plt.show()
```
